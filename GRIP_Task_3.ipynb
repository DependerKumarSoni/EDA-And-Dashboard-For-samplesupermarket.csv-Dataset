### Now we'll visualize relations between our data

```python
# Let's print the correlation among data
data.corr()
```

```python
# plotting Correlations
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(data, diag_kws=dict(fill=False))
plt.show()
```

### As per above graphs, Profit does not show any significant dependency over Postal Codes

```python
# Dropping Postal Codes
col = ['Postal Code']
data_1 = data.drop(columns=col, axis=1)
data_1.columns
```

We can see we have successfully removed 'Postal Code' from our dataframe.

### Let's Visualize our new data with a bar graph

```python
# Plotting bar graphs
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=data_1)
plt.show()
```

```python
# Let's Plot no. of entries for different states
print(data_1['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=data_1['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
# Let's have a look into Sub-Categories
print(data_1['Sub-Category'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=data_1['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

### Now Plotting the Heat Map for correlation among columns of new dataframe

```python
# Plotting correlation heatmap
fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(data_1.corr(), annot= True)
plt.show()
```

This correlation heatmap clearly tells that sales are highly correlated with Profit and well corellated by Quantity.
Well, it is obvious that Quantity will increase with sales.
The point to be noted is that, there is a negative correlation between discount & profit and discount & sales.

### Now Plotting the Heat Map for convolution among columns of new dataframe

```python
# Plotting convolution heatmap
fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(data_1.cov(), annot= True)
plt.show()
```

### Now have a look at Segment and Region Columns

```python
sns.countplot(x=data_1['Segment'])
plt.show()
```

This countplot depicts that major segments of customers include Consumers and minor segment comprises of Home Office

```python
sns.countplot(x=data_1['Region'])
plt.show()
```

This countplot show that maximum number of sales come from Western Region followed by Eastern\
region, Central region and Southern region.

### Checking how profit depends on Sub-Category.

```python
plt.figure(figsize=(40,25))
sns.barplot(x=data_1['Sub-Category'], y=data_1['Profit'])
plt.show()
```

## As we looked above, there is some correlation among Profit, Quantity and Sales.
### Let's draw some line plots to understand

```python
# Plotting lineplot between Profit and Quantity.
plt.figure(figsize = (10,4))
sns.lineplot('Quantity', 'Profit', data = data_1, color = 'g', label= 'Profit Vs Quantity')
plt.legend()
```

The graph shows a constant increment in Profit with increase in Quantity

```python
# Plotting lineplot between Profit and Sales.
plt.figure(figsize = (10,4))
sns.lineplot('Sales', 'Profit', data = data_1, color = 'r', label= 'Profit Vs Sales')
plt.legend()
```

The graph shows a sudden downfall in profit after depicting a constant increment with increasing sales

```python
# Plotting lineplot between Profit and Discount.
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Profit', data = data_1, color = 'b', label= 'Profit Vs Discount')
plt.legend()
```

## Conclusion

One thing to be noted is that while the superstore is incurring losses due to giving discounts on its products, they can't stop giving discounts of their products. Most of the heavy discounts are during festivals, end-of-season and clearance sales which are necessary so that the store can make space in their warehouses for fresh stock. Also, by incurring small losses, the company gains in the future by attracting more long term customers. Therefore, the small losses from discounts are an essential part of company's business

# The Dashboard!

```python
import plotly.express as px
import plotly.figure_factory as ff
import plotly.graph_objects as go
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output


# Load Data
df = data_1

statelist = df['State'].unique()
statelist.sort()

# Build App
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H1("Dashboard for Performing EDA on samplesuperstore.csv dataset"),
    dcc.Graph(id='graph'),
    html.Label([
        "State",
        dcc.Dropdown(
            id='state-dropdown', clearable=False,
            value='Alabama', options=[
                {'label': c, 'value': c}
                for c in statelist
            ])
    ]),
])


# Define callback to update graph
@app.callback(
    Output('graph', 'figure'),
    [Input("state-dropdown", "value")]
)

def update_figure(state):
    filtered_df = df[df['State'] == state]

    fig = px.bar(filtered_df, x='City', y='Profit')

    return fig


if __name__ == '__main__':
    app.run_server(mode='inline')
```
